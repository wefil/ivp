```python
def hist_thres(image, thresh):
    im = image.copy()  # Make a copy of the image to avoid modifying the original
    thresholded = np.where(im < thresh, 0, 255)
    return thresholded
```

```python
def binary_image(image, threshold=0.5):
    binary = np.zeros_like(image)
    binary[image > threshold] = 1
    return binary
```

```python
def calculate_iou(gt_image, pred_mask):
    gt_binary = binary_image(gt_image)
    pred_binary = binary_image(pred_mask)
    intersection = np.logical_and(gt_binary, pred_binary)
    union = np.logical_or(gt_binary, pred_binary)
    iou = np.sum(intersection) / np.sum(union)
    return iou
```

```python
def calculate_dice(gt_image, pred_mask):
    gt_binary = binary_image(gt_image)
    pred_binary = binary_image(pred_mask)
    intersection = np.logical_and(gt_binary, pred_binary)
    dice = 2 * np.sum(intersection) / (np.sum(gt_binary) + np.sum(pred_binary))
    return dice
```

```python
iou = calculate_iou(seg,k)
dice_score = calculate_dice(seg,k)

print("IoU:", iou)
print("Dice Score:", dice_score)
```
